```python
import pandas as pd
```

```python
df = pd.read_json('/Users/leenaomrani/Desktop/final-project/jobs_intelligence_artficielle_france/jobs_intelligence_france.json')
df.head()
```

```python
df['type_contrat'] = df['type_contrat'].str.replace('\n',' ')
df
```

```python
#### Convertir la colonne qui comporte des lists en string
df['titre_job2'] = [''.join(map(str, j)) for j in df['titre_job']]
df
```

```python
df['titre_job2'] = df['titre_job2'].str.replace('\n',' ')
df
```

```python
#### Convertir la colonne qui comporte des lists en string
df['date_publication'] = [''.join(map(str, d)) for d in df['date_de_publication']]
df
```

```python
df['date_publication'] = df['date_publication'].str.replace('\n',' ')
df
```

```python
df['url'] = df['url'].apply(lambda x :'https://www.hellowork.com' + x)
df
```

```python
df['duree_du_contrat'] = [''.join(map(str, j)) for j in df['duree_du_contrat']]
df['tag recruteur_reactif'] = [''.join(map(str, j)) for j in df['tag recruteur_reactif']]
df['modalites_de_tetetravail'] = [''.join(map(str, j)) for j in df['modalites_de_tetetravail']]
df
```

```python
df2=df
df2=df2.drop(axis=1, columns=['titre_job','date_de_publication'])
df2
```

```python
df2=df2.rename(columns={'titre_job2': "titre_job"})
df2
```

```python
#supprimer les doublons
result_df = df2.drop_duplicates(keep='last')
print(result_df)
```
