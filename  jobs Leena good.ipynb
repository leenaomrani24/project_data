```python
#top 25 programmin languages  
Languages_de_programmation=['Python', 'SQL', 'R', 'Java', 'Scala', 'NoSQL', 'SAS', 'Go', 'MongoDB', 'Shell', 'JavaScript', 'C++','C#','MATLAB', 'VBA', 'Bash','T-SQL','PowerShell','HTML','Perl','Ruby','TypeScript','CSS']
#top 25
Outils=['Tableau','Excel','Power BI','SAS','Word','PowerPoint','SAP','Dataiku', 'SSIS', 'Alteryx', 'SPSS', 'SRSS', 'Outlook', 'SharePoint', 'Splunk', 'MicroStrategy','DAX','Cognos','Visio','Google Sheets','Spreadsheets','MS Access','DataRobot','power bi','Qlikview']
#top25
Cloud=['AWS','Azure','Snowflake','Redshift','DataBricks','Oracle','Google Cloud Platform','BigQuery','Aurora','IBM Cloud','VMware','Watson','Firebase','OpenStack','Heroku','Digital Ocean ','Colocation','OVH','Linode']
#top25
Librairies=['Spark','Hadoop','Kafka', 'Airflow', 'PySpark', 'Pandas', 'TensorFlow','NumPy','PyTorch','Scikit-learn','Keras','Jupyter','React', 'Matplotlib','Spring','Plotly','GraphQL','Seaborn','OpenCV','NLTK','ggplot2','Selenium','Tidyverse','MXNet']
#top 25 
Base_de_données=['SQL Server','MySQL','Cassandra','PostgreSQL','MongoDB','Elasticsearch','DynamoDB','DB2','Redis','Neo4j','MariaDB','Firebase','Couchbase','Firestore','CouchDB','SQLite','sql server']

#top 25 

Autres=['RGPD','Data Privacy','Data Quality','Data Gouvernance','Data Management','Data Protection','Data Compliance']
```

```python
def categorize_skills(Compétences):
    skill_categories = {'Languages_de_programmation': [], 'Outils': [], 'Cloud': [], 'Librairies': [], 'Base_de_données': [],'Autres':[]}
    for skill in Compétences:
        if skill in Languages_de_programmation :
            skill_categories['Languages_de_programmation'].append(skill)
        elif skill in Outils:
            skill_categories['Outils'].append(skill)
        elif skill in Cloud:
            skill_categories['Cloud'].append(skill)
        elif skill in Librairies:
            skill_categories['Librairies'].append(skill)
        elif skill in Base_de_données:
            skill_categories['Base_de_données'].append(skill)
        elif skill in Autres :
            skill_categories['Autres'].append(skill)
    return pd.Series(skill_categories)

# Apply the custom function to the skills column
df1[['Languages_de_programmation', 'Outils', 'Cloud', 'Librairies', 'Base_de_données','Autres']] = df1['Compétences'].apply(categorize_skills)
```
