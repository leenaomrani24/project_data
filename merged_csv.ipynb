```python
import re

def remove_arrondissement(ville_str):
    pattern = r"\b(Paris\s(?:[1-9]|1[0-9]|20))(?:er)?\b"
    match = re.search(pattern, ville_str)
    if match:
        return match.group(1)
    else:
        return ville_str

df['Ville'] = df['Ville'].apply(remove_arrondissement)
```

```python
import re

df['Ville'] = df['Ville'].apply(lambda x: re.sub(r'\d+', '', x))
```

```python
import re
# take the words in code postale column 

# Define a regular expression pattern to match any non-digit characters
pattern = r'\D'

# Apply the regular expression pattern to the Code Postale column and replace with an empty string
df['Code Postale'] = df['Code Postale'].apply(lambda x: re.sub(pattern, '', str(x)))
```

```python
from datetime import datetime, timedelta
import dateparser

def replace_date(date_str):
    if date_str == "hier":
        return (datetime.datetime.today() - datetime.timedelta(days=1)).strftime('%d/%m/%Y')
    elif date_str.startswith("il y a"):
        hours = int(date_str.split(" ")[2])
        return (datetime.datetime.now() - datetime.timedelta(hours=hours)).strftime('%d/%m/%Y')
    elif not date_str.strip(): # check if date_str is an empty string or whitespace only
        return "" # return empty string if date_str is invalid
    else:
        return dateparser.parse(date_str).strftime('%d/%m/%Y')

df['date_publication'] = df['date_publication'].apply(replace_date)
```
