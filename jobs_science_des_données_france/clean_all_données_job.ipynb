```python
def extract_id_offre(url_offre):
    match = re.search(r'/(\d+)\.html', url_offre)
    if match:
        return match.group(1)
    else:
        return None

# apply the function to the 'url' column and create a new 'number' column
df['id_offre'] = df['url_offre'].apply(extract_id_offre)

# print the resulting DataFrame
print(df)
```
