```python
import pandas as pd
```

```python
df=pd.read_json('HW_Jobs_data_France_details.json')
df
```

```python
import re
```

```python
df['descriptif_job'] = [''.join(map(str, d)) for d in df['descriptif_job']]
df['tags'] = [''.join(map(str, d)) for d in df['tags']]
df
```

```python
df['descriptif_job'] = df['descriptif_job'].str.replace('\n',' ')
df
```

```python
df['descriptif_job'] = df['descriptif_job'].str.replace('\t',' ')
df
```

```python
df['url_offre']
```

```python
def extract_id_offre(url_offre):
    match = re.search(r'/(\d+)\.html', url_offre)
    if match:
        return match.group(1)
    else:
        return None

# apply the function to the 'url' column and create a new 'number' column
df['id_offre'] = df['url_offre'].apply(extract_id_offre)

# print the resulting DataFrame
print(df)
```
